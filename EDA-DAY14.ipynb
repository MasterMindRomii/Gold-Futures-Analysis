# Data Visualization

```python
plt.figure(figsize=(12,8))
sns.histplot(x='Price',data=df,y='Date')
plt.title('Distribution of Data Over Time')
plt.xlabel('High')
plt.ylabel('Date')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x='Price',data=df,y='Date')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.show()
```

# Refrence Visualization

```python
# Create a time series plot for 'Price', 'Open', 'High', and 'Low' over time
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'], label='Price')
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title("Price, Open, High, and Low Over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create a histogram for the 'Date' column
plt.figure(figsize=(8, 6))
plt.hist(df['Date'], bins=30, edgecolor='k')
plt.xlabel('Date')
plt.ylabel('Frequency')
plt.title("Distribution of Data Over Time")
plt.grid(True)
plt.show()
```

```python
# Create a line chart for 'Date' and 'Price'
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Price Trends Over Time')
plt.grid(True)
plt.show()
```

```python
# Create a line chart for 'Date' and 'High'
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['High'])
plt.xlabel('Date')
plt.ylabel('High Price')
plt.title('High Price Trends Over Time')
plt.grid(True)
plt.show()
```

```python
# Create a line chart for 'Date' and 'Low'
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Low'])
plt.xlabel('Date')
plt.ylabel('Low Price')
plt.title('Low Price Trends Over Time')
plt.grid(True)
plt.show()
```

# Advance Visualization

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Create a pair plot
sns.pairplot(df)
plt.show()
```

```python
# Create a violin plot for 'Price'
plt.figure(figsize=(10, 6))
sns.violinplot(x='Price', data=df, palette='Set3')
plt.title('Distribution of Price (Violin Plot)')
plt.show()
```

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Open'], df['High'], df['Low'])
ax.set_xlabel('Open Price')
ax.set_ylabel('High Price')
ax.set_zlabel('Low Price')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# Create a KDE plot for 'Price'
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Price'], shade=True)
plt.title('Price Distribution with KDE')
plt.show()
```
